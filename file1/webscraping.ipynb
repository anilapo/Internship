```python
import requests
from bs4 import BeautifulSoup

url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

articles = soup.find_all("div", class_="Card-titleContainer")

for article in articles:
    heading = article.find("a", class_="Card-titleLink")
    date = article.find("time", class_="Card-time")
    link = article.find("a", class_="Card-titleLink")

    if heading:
        heading_text = heading.text.strip()
    else:
        heading_text = "Not Available"

    if date:
        date_text = date.text.strip()
    else:
        date_text = "Not Available"

    if link:
        link_href = "https://www.cnbc.com" + link["href"]
    else:
        link_href = "Not Available"

    print("Heading:", heading_text)
    print("Date:", date_text)
    print("News Link:", link_href)
    print("\n")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Check if the article list is found before proceeding
article_list = soup.find("div", class_="article-list")
if article_list:
    # Extracting the most downloaded article details
    article = article_list.find("div", class_="article-item")
    
    if article:
        # Extracting paper title, date, and author
        paper_title = article.find("h3").text.strip()
        date = article.find("p").text.split()[-1]
        author = article.find("p").text.split("Share article")[0].strip()

        print("Paper Title:", paper_title)
        print("Date:", date)
        print("Author:", author)
    else:
        print("No article details found.")
else:
    print("Article list not found on the page.")
```
